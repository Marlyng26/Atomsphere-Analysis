```python
#Preparing the Data Unlike the simple LR, this time we are going to use column names for creating an attribute set and label.
X=df[['pop','income','ipaddr','ufo2010']]
y=df['infections']
print(X.shape)
print(y.shape)
```

```python
#pratice preparing the data in x and y variable
X=df[['pop','income','ipaddr','ufo2010']]
y=df['infections']
print(X)
print(y)
```

```python
#Dividing our data into training and test sets:
#the test_size is indicating the size of the percentage of testing
#70/30 Split: Mostly used for datasets that are resonably sized.
#80/20 Split: Typically used for larger datasets where a larger training set is constructive.
#90/10 Split: At times used for very large datasets to safegaurd enough training data.
#random_state establishes duplication by setting a seed for random number generation.
# In order to test and train we need to import train_test_split from sklearn.model_selection

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

```

```python
#Training the Algorithm
# in order to create linear regressions we need to import LinearRegression from sklearn.linear_model
from sklearn.linear_model import LinearRegression
#reg = LinearRgression() creates an instance of the LinearRegression class and assigns it to the variable reg.
reg=LinearRegression()
#the reg.fit()  results in the fit method on the reg instance and passing in X_train and y_train from the data.
reg.fit(X_train,y_train)
```

```python
#In case of multivariable linear regression, the regression model has to find the most optimal coefficients for all the attributes.
coeff_df = pd.DataFrame(reg.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
#In case of multivariable linear regression, the regression model has to find the most optimal coefficients for all the attributes.
#for y= number of infections
#y=0.002363 pop + 0.048289 income -0.002468 ipaddr + 78.018269 ufo2010
```

```python
#Making Predictions
#In reg.pred(X_test) reg refers to a trained regresion model. The X_test is the set of independent variables from the datset that the
#model will use to make predictions. The method predict() is used to predict on new data based on the patterns the model has learned
#during the training
y_pred=reg.predict(X_test)
y_pred
```

#comparing the actual output values for X_test with the predicted values:
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df


```python
#Evaluating the Algorithm
# Error metrics
# Using the following will result in printing the Abosulte error, mean sqaure error, and root mean sqaured error.
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#The metrics for the function to assess a regression model's performance, the R-squared (R²) score is computed using the formula r2_score(y_test, y_pred).
#The y_test is the true values of the dependent variable from the test dataset. y_pred is the predicted value generated by the model for the corresponding X_test
# The R sqaured is a statistical measure that demonstrates how well the predictions of a regression model appraoximate the real data points. It givves us the ratio fo variance
#in the dependent variable that is explained by the independent variable in the model.
metrics.r2_score(y_test,y_pred)
```

```python
#We use this line of code to select subsets of columns from the DataFrame to serve as independent variables for machine learning
#X1 contains the independent variables
#y1 contains the dependent variable that we are trying to predict
X1=df[['pop','income','ufo2010']]
y1=df['infections']
```

```python
#train_test_split method divides the dataset into training and testing sets
#X1_train and y1_train are used to train the model similarily, X1_test and y1_test are
#used to test the model. The test_size = .2 means that 20 % of the data will be used for testing
#and 80 % for trianing. The random_state  function will splot the data the same way every time we
#run the code. If we do not set a specific value for random_state, then the function will split
#the data differently each time the code is run
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=0)
```

```python
reg=LinearRegression()
reg.fit(X1_train,y1_train)
```

```python
# The following creates a DataFrame that summarizes the coefficients of a fitted linear regression model in a organized manner
coeff_df = pd.DataFrame(reg.coef_, X1.columns, columns=['Coefficient'])
coeff_df
```

```python
#Practice creating summary of coeffcients of DataFrame
# This is used to create a df that consists of the coefficients of a trained
#regression model along with the names of the corresponding features fom X1.columns.
#the reg.coef is an attribute of a trained regression model.
#X1_columns refers to teh column names of the X1 df which contains the independent vvaribales used to
#train the model. The whole line of code creates a new Pandas DataFrame from the coefficients in reg.coef and labels
#row with features name from X1.columns
coeff_df = pd.DataFrame(reg.coef_, X1.columns, columns=['Coefficient'])
coeff_df
```

```python
#Making Predictions
#Here we are storng the previsouly treained model into y1_pred.
#The reg portion of the code pertains to to the previously trained model
#predict() is the method used to make predictions with the trained model
#(X1_test) containes the independent variables of the test set of the data
y1_pred=reg.predict(X1_test)
```

```python
#comparing the actual output values for X_test with the predicted values:
#Thsi line of code displays the y1_test and y_1pred
df=pd.DataFrame({'Actual':y1_test, 'Predicted':y1_pred})
df
```

```python
#Evaluating the Algorithm
# Here we are importing the metrics model from sklearn whcih includes numerous functions for
#evaluating machine leanring models.
# Error metrics
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y1_test, y1_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y1_test, y1_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y1_test, y1_pred)))
```

```python
# Calucaltes the R sqaured score for the regression model's predictions. The R sqaure is a measure
#of how well the predcited values match the actual values. This provides insight into how well
#the model explains the variability of the target variables.
metrics.r2_score(y1_test,y1_pred)
```
